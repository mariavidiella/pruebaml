```python
import os
```

```python
ROUTE = "train_data/source/"
```

```python
dirs_files = [os.path.join(ROUTE, f) for f in os.listdir(ROUTE) if f.endswith('.csv')]
```

```python
dirs_files
```

```python
dfs = []

for file in dirs_files:
    df = pd.read_csv(file)
    city_name1 = file.split('/')[-1]
    city_name = city_name1.split('.csv')[0]
    df.insert(1,"city", city_name)
    dfs.append(df)
```

```python
[df.shape for df in dfs]
```

```python
df_combined = pd.concat(dfs, ignore_index= True)
```

```python
df_combined
```

```python
df_combined.shape
```

```python
df_combined = df_combined.sample(frac = 1) # Para reordenar el df de modo que no este ordenado, frac = 1 porque cojo todos los datos
```

```python
df_combined.head()
```

```python
df_combined.to_csv("train_data/preprocessed/combined_data.csv", index=False)
```
